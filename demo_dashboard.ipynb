```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output
from mvsm_model import MVSMEngine
```

## Parameter Setup

Configure the model parameters below:

```python
# Initialize parameters
p_start = 112      # Initial price
gap_init = 1.5     # Initial gap
intensity = 0.3    # Event intensity
stock_init = 45    # Initial stock

print(f"Model Parameters:")
print(f"  Initial Price: ${p_start}")
print(f"  Initial Gap: {gap_init} mb/d")
print(f"  Event Intensity: {intensity}")
print(f"  Initial Stock: {stock_init} days")
```

## Run Simulation

Execute the model with 5,000 Monte Carlo paths:

```python
# Run the engine
engine = MVSMEngine(p_start, gap_init, intensity, stock_init)
prices = engine.simulate()

print(f"Simulation complete: {len(prices)} paths generated")
print(f"\nPrice Statistics:")
print(f"  Mean: ${np.mean(prices):.2f}")
print(f"  Median: ${np.median(prices):.2f}")
print(f"  Std Dev: ${np.std(prices):.2f}")
print(f"  Min: ${np.min(prices):.2f}")
print(f"  Max: ${np.max(prices):.2f}")
```

## Results Visualization

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.hist(prices, bins=60, color='#2c3e50', alpha=0.8, density=True, edgecolor='black')
ax.axvspan(150, max(prices), color='red', alpha=0.1, label='Rupture Zone (>$150)')
ax.axvline(np.mean(prices), color='green', linestyle='--', linewidth=2, label=f'Mean: ${np.mean(prices):.2f}')
ax.set_title("MVSM 12.0: Monte Carlo Price Distribution @ March 28", fontsize=14, fontweight='bold')
ax.set_xlabel("Price ($)", fontsize=12)
ax.set_ylabel("Density", fontsize=12)
ax.legend(fontsize=10)
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

## Risk Analysis

```python
# Calculate key risk metrics
rupture_threshold = 150
prob_rupture = np.sum(prices > rupture_threshold) / len(prices)

print(f"Risk Metrics:")
print(f"  Logic Rupture Probability (>${rupture_threshold}): {prob_rupture:.2%}")
print(f"  Number of Rupture Scenarios: {int(np.sum(prices > rupture_threshold))} / {len(prices)}")
print(f"\n  VaR (95%): ${np.percentile(prices, 95):.2f}")
print(f"  CVaR (95%): ${np.mean(prices[prices > np.percentile(prices, 95)]):.2f}")
```
